```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from statsmodels.tsa.stattools import adfuller
import numpy as np

# List of CSV files
csv_files = [
    "dry blub temperature.csv",
    "potential temperature.csv",
    "sea surface temperature.csv",
    "surface heigh.csv",
    "surface_height.csv",
    "vertical velocity at t points.csv"
]

# Dictionary to store DataFrames
dfs = {}

# Load datasets
for file in csv_files:
    if os.path.exists(file):
        try:
            df = pd.read_csv(file, delimiter=",", encoding="utf-8", on_bad_lines="skip", engine="python", skiprows=8)  # Skip metadata

            # Convert all columns (except datetime) to numeric, forcing errors to NaN
            for col in df.columns:
                if col not in ["DATETIME", "TIME"]:  # Exclude time-related columns
                    df[col] = pd.to_numeric(df[col], errors="coerce")

            # Replace -1.E+34 (error values) with NaN
            df.replace([-1e+34, "-1.E+34", -1.0e+34, "-1.000000e+34"], np.nan, inplace=True)

            dfs[file] = df
            print(f"‚úÖ Loaded {file} successfully! Shape: {df.shape}")
        except Exception as e:
            print(f"‚ùå Error loading {file}: {e}")
    else:
        print(f"‚ö†Ô∏è Warning: {file} not found!")

# Function to check preprocessing
def check_preprocessing(dfs):
    for file, df in dfs.items():
        print(f"\nüìÇ **Checking {file}**")
        
        # 1Ô∏è‚É£ Check for missing values
        missing_values = df.isnull().sum().sum()
        print(f"üîç Missing Values: {missing_values}")
        
        # 2Ô∏è‚É£ Check for duplicate rows
        duplicates = df.duplicated().sum()
        print(f"üîç Duplicate Rows: {duplicates}")

        # 3Ô∏è‚É£ Check data types
        print("üîç Data Types:")
        print(df.dtypes)

        # 4Ô∏è‚É£ Check for outliers (visualization)
        numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns
        if not numeric_cols.empty:
            plt.figure(figsize=(10, 5))
            sns.boxplot(data=df[numeric_cols])
            plt.title(f'Outlier Check - {file}')
            plt.show()

        # 5Ô∏è‚É£ Check MinMax Scaling (only for numeric columns)
        if not numeric_cols.empty:
            scaler = MinMaxScaler()
            scaled_data = scaler.fit_transform(df[numeric_cols])
            print(f"üîç Scaling Check: Min={scaled_data.min():.2f}, Max={scaled_data.max():.2f}")

        # 6Ô∏è‚É£ Check time-series continuity
        if 'DATETIME' in df.columns:
            # Try multiple formats
            for fmt in ["%d-%b-%Y", "%Y-%m-%d", "%d/%m/%Y", "%m-%d-%Y"]:
                try:
                    df['DATETIME'] = pd.to_datetime(df['DATETIME'], format=fmt, errors='coerce')
                    if not df['DATETIME'].isna().all():
                        break  # Stop if successful
                except:
                    continue
            
            missing_dates = pd.date_range(start=df['DATETIME'].min(), end=df['DATETIME'].max(), freq='D').difference(df['DATETIME'])
            print(f"üîç Missing Time Steps: {len(missing_dates)}")

        # 7Ô∏è‚É£ Check for stationarity (ADF Test) if SST column exists
        if 'SST' in df.columns:
            try:
                result = adfuller(df['SST'].dropna())
                print(f"üìä ADF Statistic: {result[0]:.2f}, p-value: {result[1]:.4f}")
                print("‚úÖ Stationary" if result[1] < 0.05 else "‚ö†Ô∏è Non-stationary (consider differencing)")
            except Exception as e:
                print(f"‚ö†Ô∏è ADF test error: {e}")

# Run preprocessing checks
check_preprocessing(dfs)

# Function to fill missing values and print before/after status
def fill_missing_values(dfs):
    for file, df in dfs.items():
        print(f"\nüìÇ **Checking Missing Values in {file}**")
        
        # Before filling missing values
        missing_before = df.isnull().sum().sum()
        print(f"üîç Missing Values Before Filling: {missing_before}")
        
        # 1Ô∏è‚É£ Fill missing values using bfill, ffill, and interpolation
        df.bfill(axis=0, inplace=True)  # Backfill
        df.ffill(axis=0, inplace=True)  # Forward fill
        df.interpolate(method='linear', axis=0, inplace=True)  # Linear interpolation
        
        # After filling missing values
        missing_after = df.isnull().sum().sum()
        print(f"üîç Missing Values After Filling: {missing_after}")


# Fill missing values
fill_missing_values(dfs)

# Print the first 20 rows of each dataset after filling missing values
for file, df in dfs.items():
    print(f"\nüìÇ First 20 rows of {file} after filling missing values:")
    print(df.head(20))

# Total number of rows across all datasets
total_rows = sum(df.shape[0] for df in dfs.values())
print(f"üìä Total number of rows across all datasets: {total_rows}")
```
